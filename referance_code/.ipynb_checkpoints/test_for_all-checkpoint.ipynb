```python
import pandas as pd

#data open using pandas type Dataframe
df_standard = pd.read_csv ('fd_basic_F.csv', encoding = 'cp949')
df_to_add = pd.read_csv ('fd_2018_F.csv', encoding = 'cp949')

df_samsung_0 = df_standard[df_standard['code']==5930]
df_samsung_1 = df_to_add[df_to_add['code']==5930]

#Extract key value from df_to_add
df_to_add_key = df_samsung_1[['y2017','y2018']]

#df_to_add_listing
df_to_add_list =[]
for i in range(len(df_samsung_1)):
    data_2017 = df_to_add_key.iloc[i,0]
    data_2018 = df_to_add_key.iloc[i,1]
    df_to_add_list.append([data_2017,data_2018])

#df_standard_listing
df_standard_list = []
for i in range(len(df_samsung_0)):
    index_data_2017 = df_samsung_0.iloc[i,6]
    df_standard_list.append(index_data_2017)

#merge data
for i in range(len(df_standard_list)):
    for j in range(len(df_to_add_list)):
        index_data = df_standard_list[i]
        if index_data == df_to_add_list[j][0] : df_standard_list[i] = [index_data, df_to_add_list[j][1]]

#non data make 0    
for i in range(len(df_standard_list)):
    non_data = df_standard_list[i]
    if type(df_standard_list[i]) != list : df_standard_list[i] = [non_data,0]

#make index column
for i in range(len(df_standard_list)):
    df_standard_list[i].insert(0,df_samsung_0.iloc[i,0])

    


indexs=[]
y2017s=[]
y2018s=[]
for i in range(len(df_standard_list)):
    index = df_standard_list[i][0]
    y2017 = df_standard_list[i][1]
    y2018 = df_standard_list[i][2]
    indexs.append(index)
    y2017s.append(y2017)
    y2018s.append(y2018)
a = pd.DataFrame({'Unnamed: 0':indexs,'y2018':y2018s})
merged_inner = pd.merge(left=df_samsung_0, right=a, left_on='Unnamed: 0', right_on='Unnamed: 0')
merged_inner
```
